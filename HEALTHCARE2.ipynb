```python
#The name column is not a correct order as all the names are in upper case letter, we have to correct them
df['Name']=df['Name'].str.title()
```

```python
#There are some names with un certain dots and commas at the end of names and they need to be remove
df['Name']=df['Name'].str.rstrip('.')
```

```python
#Changing the datatype of the Date of admission Column and Discharge-Date
'''
AS the column is not in the right data type and can cause issues when further moving in with the data it's better to change the data type first 
'''
df['Date of Admission']=pd.to_datetime(df['Date of Admission'], format='%Y-%m-%d')
```

```python
#Changing the datatype of the Discharge Date
df['Discharge Date']=pd.to_datetime(df['Discharge Date'],format="%Y-%m-%d")
```

```python
'''
Starting the data analysis and perofroming statistical analysis accompnied with visualisation in order to understand the trends from the data.
'''
#Medical Condition
df.head(2)
```

```python
#1.Finding the medical condition by age using the boxplot
plt.figure(figsize=(10,5))
plt.style.use('bmh')
sns.boxplot(data=df, x="Medical Condition", y='Age', palette="Set3")
plt.title("Age Distribution By Medical Condition")
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
'''
using a box plot to know what age was mostly affected with medical condition
'''
```

```python
'''
Now we will see the Medical Condition by Gender, which will help us understand which Gender was mostly effected by the medical condition
'''
medicalgender=df.groupby(["Medical Condition", "Gender"]).size().reset_index(name='Counts')

medicalgender
```

```python
'''
Using a barplot for the visualizing medical condition sideby side of their gender, this will easily help us understand how medical condition was in the
respective gender
'''
#Medical Condition by gender using a count plot
plt.figure(figsize=(16,5))
plt.style.use("fivethirtyeight")
sns.set_context("paper", font_scale=1.8)
ax=sns.barplot(data=medicalgender, x="Medical Condition", y="Counts", hue="Gender", palette="Set2")
plt.title("Medical Condition By Gender")
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=3)
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
```

```python
#Finding the Hospital and Insurance analysis
#Finding the top 10 hospital in which the pateints went
plt.figure(figsize=(10,5))
df["Hospital"].value_counts().nlargest(10).plot(kind='bar',cmap='magma')
plt.title("Top 10 Hospital")
plt.xlabel("Hospital")
plt.ylabel('Patients Counts')
plt.xticks(rotation=30)
```

```python
'''
Finding the Average bill amount of the insrance provider by using a bar plot, this will help the patient know which insurance covers most of the cost and
this help them chose whats good for them
'''
billinginsurance=df.groupby("Insurance Provider")['Billing Amount'].mean().reset_index()
billinginsurance
```

```python
'''
This bar plot will help us know the difference between the cost covered by the respective insurance providers
'''
#Now we will use a bar plot by seabron to visualize it
plt.figure(figsize=(20,8))
plt.style.use("bmh")
sns.set_context('talk', font_scale=1)
ax=sns.barplot(data=billinginsurance, x="Insurance Provider", y="Billing Amount", palette="muted")
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=2)
plt.title("Average billing Amount by Insurance Provider")
plt.xlabel("Insurance Provider")
plt.ylabel("Average Billing Amount ($)")
```

```python
'''
Using the 'Date OF Admission' column which was previously changed to datetime column, we will pull out the months from it by using .dt.month query and
see the months where most of the patients were seen overall
'''

df['Months']=df["Date of Admission"].dt.month

# Now we have a new column with number of months , this will help us understand what sort of admssions were done through out 12 months of the year
plt.figure(figsize=(14,6))
plt.style.use("fivethirtyeight")
sns.set_context('talk')
ax=sns.countplot(df, x='Months', hue="Admission Type", palette="colorblind")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
```

```python
#Finding the length of the stay of each patient
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
plt.figure(figsize=(10,6))
sns.histplot(df['Length of Stay'], bins=20, kde=True, palette="magma")
plt.title('Distribution of Length of Stay (Days)')
plt.xlabel('Days')
plt.ylabel('Count')
plt.show()
```

```python
'''
From the above graph i will now use the column length of stay with the medical condition as it will give us the exact numbers of days each patients were
in the hospital depending upon their days of medical condition.
'''
aggregate_stats = df.groupby('Medical Condition')['Length of Stay'].agg(
    Count='count',
    Mean='mean',
    Median='median',
    Min='min',
    Max='max',
    Std_Dev='std'
).reset_index()
aggregate_stats
```

```python
#Finding the length of stay by medical condition
plt.figure(figsize=(15,7))
plt.style.use("bmh")
sns.set_context("paper", font_scale=1)
sns.boxplot(data=df, x="Medical Condition", y ="Length of Stay", palette="Set2")
```

```python
#Finding the  Average billing amount by the type of admission
billing_admission=df.groupby("Admission Type")["Billing Amount"].mean().reset_index()
billing_admission
#Using a bar plot to plot the grapgh to understand the average billing by admission type
```

```python
#Using a bar plot to plot the grapgh to understand the average billing by admission type
plt.figure(figsize=(15,7))
plt.style.use('bmh')
sns.set_context("paper", font_scale=1.6)
ax=sns.barplot(billing_admission, x="Admission Type", y="Billing Amount", palette="Set2")
plt.title("Billing Amount By Admission Type")
plt.ylabel("Billing Amount ($)")
plt.tight_layout()

#Adding values on the top of the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', padding=1)
plt.show()
```

```python
#Finding the top 10 Medication prescribed by the doctor to the patients although there are total of 5 Medicines
plt.figure(figsize=(12,4))
plt.style.use("fivethirtyeight")
sns.set_context(font_scale=0.5)
df['Medication'].value_counts().nlargest(10).plot(kind='bar', color='seagreen')
plt.title("Top 10 Medications")
plt.ylabel("Counts")

```

```python
'''
In order to understand the how the medicines were given to patients depending upong their medical condition, i will make a piviot table.
This will allow us to see the total number of medicines with the type given to certain meidcal conditon.
Lastly we can use that Pivot table as out heatmap aswell
'''

prescription=df.groupby(['Medical Condition', "Medication"]).size().unstack()
prescription
```

```python
#Through this heatmap we can clearly see which medicine was prescibed for what sort of medical condition.
#The darker the color the more that stuff is being used, althoug we can see the line defining the color trends easily for us
plt.figure(figsize=(15,7))
sns.heatmap(prescription, annot=True, cmap='YlGnBu', fmt='0.2f')
plt.title("Presciption Based on Medical Condition")
plt.xlabel("Medicines")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
'''
Making a pie chart to show the Test Result, it will help us understand what were the outcomes of the medical conditions and how were the patients
after their medication as this could help hospital, doctors and in the improvement of the meidcations.
'''
#Test Result Analysis
#Pie chart will differntiate how the test results were, meaning how did the recoveries went.
plt.figure(figsize=(10,4))
df['Test Results'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Test Result Distribution')
plt.ylabel('')
plt.show()
```

```python
'''
Finding the test reult based on the Medical condition, with this we will be able to know about the risks one medical condition had and how were the
result of different medical-conditions

'''
#Using a count plot
plt.figure(figsize=(18,6))
plt.style.use("ggplot")
sns.set_context('paper',font_scale=1.5)
ax=sns.countplot(df, x="Test Results", hue="Medical Condition", palette='Set3')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=1)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.xlabel("Test-Results")
plt.ylabel('Counts')
plt.xticks(rotation=40)
plt.tight_layout()
```
